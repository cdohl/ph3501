# Hydrostatics


In hydrostatics we do not consider forces due to liquid motion. The Cauchy Equation of motion becomes therefore very simple. Let's start with the original equation:

$$\rho\fudm{u_j}{t}=\rho g_j +\pad{}{x_i}\tau_{ij}\tag{1}$$

The L.H.S. of Eq. (1) is 0 in hydrostatics, and on the R.H.S. we only account for forces caused by a stagnant liquid. In particular, we account only for nomal forces and write the stress tensor as

$$\tau_{ij}=-p\,I_{ij}\tag{2}\quad ,$$

where $\tb{I}$ is the indentity matrix. This way only normal forces per area are accounted for and these are termed pressure $p$. The components along the diagonal must be of equal size to balance each other, thus *not* to accelerate the fluid particle. All off diagonal elements of $\tb{\tau}$ must vanish, they would otherwise either rotate or strain the fluid elements and again accelerate.

Inserting the stress tensor Eq. (2) into Eq. (1) and considering a stagnant liquid only we obtain

$$ 0 = \rho g_j + \pad{}{x_i} \left(-p I_{ij}\right)\tag{3}\quad .$$


let's write Eq. (3) for $j=1$ to exercise a bit Einstein summation convention:

$$ 0 = \rho(\vb{x}) g_1 - \left[\pad{}{x_1}\left(p(\vb{x})\,I_{11}\right) 
+\pad{}{x_2}\left(p(\vb{x})\,I_{21}\right)
+\pad{}{x_3}\left(p(\vb{x})\,I_{31}\right)
\right]\tag{4}$$

The last two terms in the square brackets are 0 (why?) and we can write Eq. (3) in short

$$\rho\,\vb{g} = \nabla p\tag{5}\quad .$$

In general $\rho=\rho(\vb{x})$ and integration of Eq. (5) needs the equation of state for the fluid. For liquids the compressibilty, i.e. the variation of density with pressure is weak and the density of a liquid can be safely considered constant for most of the cases.

Thus for liquids we can integrate Eq. (5) and obtain

$$\rho \,\vb{g} \cdot \vb{x} + c = p\tag{6}\quad.$$

## Archimedes Principle

Archimedes principle obtains the force acting on a rigid body immersed in a stagnant fluid in presence of gravity. 

<div class="Figure"> 
<img src="files/pics/immersed_body.png" width=200> 
Figure 1: A body with normal vector $\vb{n}$ pointing outward is immersed in a liquid. Gravity is given by $\vb{g}$.
</div>

The force is obtained by integrating the pressure over the surface of the body.
$$\vb{F}_S=-\int\int_S p\, \vb{n} \dd S\tag{7}$$

The pressure is given by Eq. (6) and we can write

$$\vb{F}_S=-\int\int_S \left[\rho \left(g_1 x_1+g_2 x_2 + g_3 x_3\right) + p_0\right] \vb{n} \dd S\tag{8}$$

Applying Gauss divergence theorem to Eq. (8) we obtain

$$\vb{F}_S=-\int\int\int_V -\rho \vb{g} \dd V = -\rho \vb{g} \int\int\int \dd V = -\rho \vb{g} V$$

Thus the force is proportional to the displaced volume and density of the liquid and it is opposite to the action of gravity. This law has been named after Archimedes who stated:
