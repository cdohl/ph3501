# Dimensionless form of the Navier Stokes Equation

The Navier Stokes equation

$$\rho \left(\pad{\vb{u}}{t} + \vb{u}\cdot \nabla \vb{u}\right) =
-\nabla p + \rho \vb{g} +\mu \nabla^2\vb{u}\tag{10}$$

can be written in dimensionless form. All parameters with units will end up into dimensionless numbers characterizing the solution. We write the dimensionless number with a prime, e.g. $x'$.

We start with the length
$$x'_j=x_j/l \quad ,$$
where $l$ is a length scale, the nondimensional time
$$t'=\Omega t\quad ,$$
where $\Omega$ is an imposed frequency. The nondimensional velocity is
$$u'_j=u_j/U\quad ,$$
where $U$ is a typical velocity. For the pressure we need to identify again a
typical pressure. That depends on the problem; it could be the atmospheric pressure, interfacial pressure, or here we suggest the stagnation pressure
$$p'=\frac{p-p_0}{\rho U^2}\quad ,$$
where $p_0$ is the pressure far from the stagnation point. Non-dimensionalising the body force is trivial

$$g'_j=g_j/g\quad .$$

The N.S. equation contains a number of derivatives. We show now how to nondimensionalize these expression with an example

$$\pad{u_j}{t}=\pad{U u'_j}{\Omega^{-1}t'}=U\Omega\pad{u'_j}{t'}$$

Using this approach let's write our N.S. equation (10) in non-dimensional form:

$$\left[\frac{\Omega l}{U}\right]\pad{\vb{u'}}{t'} + \vb{u'}\cdot \nabla'
\vb{u'} =
-\nabla' p' + \left[\frac{g l}{U^2}\right]\vb{g'} +\left[\frac{\mu}{\rho U l}\right] \nabla'^2\vb{u'}\tag{11}$$



There are a number of nondimenisonal numbers introduced in literature which appear in Eq. (10). Before we try to identify them we list 3 important ones for incompressible flows.

## Reynolds number

The [Reynolds](https://en.wikipedia.org/wiki/Osborne_Reynolds) number is the ratio of the inertia to viscous force, i.e. 

$$\mathrm{Re}=\frac{\mathrm{inertia}}{\mathrm{viscous\,force}}\propto
\frac{\rho u \,(\partial u/\partial x)}{\mu \,(\partial^2u/\partial x^2)}\propto
\frac{\rho U^2/l}{\mu U/l^2}=\frac{\rho U l}{\mu}$$

## Froude number

The [Froude](https://en.wikipedia.org/wiki/William_Froude) number compares inertia with gravity force, i.e.

$$\mathrm{Fr}=\left[\frac{\mathrm{inertia}}{\mathrm{gravity\,force}}\right]^{1/2}\propto
\left[\frac{\rho u \,(\partial u/\partial x)}{\rho g}\right]^{1/2}\propto
\left[\frac{\rho U^2/l}{\rho g}\right]^{1/2}=\frac{U}{\sqrt{g l}}$$

## Strouhal number

The Strouhal number is defined as the ragtio of the time dependent acceleration (unsteady) and the geometrical acceleration (advective term)


$$\mathrm{St}=\frac{\mathrm{unsteady\,acceleration}}{\mathrm{advective\,acceleration}}
\propto\frac{\partial u/\partial t)}{u(\partial u/\partial x)}\propto
\frac{\Omega U}{U^2/l}=\frac{\Omega l}{U}$$


We can now rewrite Eq. (10) using above 3 numbers:

$$\mathrm{St}\pad{\vb{u'}}{t'} + \vb{u'}\cdot \nabla'
\vb{u'} =
-\nabla' p' +\frac{1}{\mathrm{Fr}^2}\vb{g'} +\frac{1}{\mathrm{Re}} \nabla'^2\vb{u'}\tag{12}$$

# Dynamic similarity

A flow is similar if the nondimensional parameters appearing in Eq. (11) are the same. That means, if Re, Fr, and St numbers are the same. Thus an experiment with the same non-dimensional numbers leads to the same non-dimensional solution (the dimensional solution is still different). In general it is impossible to keep all 3 dimensional numbers the same. Yet, for many flows not all of the three numbers are equally important. For example, gravity hardly is important for internal flows such as a pipe flow, or the flow around an object, while for surface gravity waves it must be considered. Also for steady flows the St number is 0 and we only have to deal with a Re number similarity.
