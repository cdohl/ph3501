# Fluid Stresses at Interfaces

## Fluid Stresses at a Rigid Boundary

Newton's 3rd law states that the action of every force comes with a force acting in the opposite direction and equal magnitude. The fish swimming through water uses it fins to accelerate the fluid in the opposite direction from its motion. Consider now a rigid boundary where a fluid acts upon, thus Newton's 3rd law states that the force of the liquid on the boundary is the same as the boundary on the fluid, but in opposite direction. Have a look at Fig. 1.

<div class="Figure"> 
<img src="files/pics/boundary.png" width=300> 
Figure 2:
</div>


The liquid stress onto the boundary is equal and opposite to the stress exerted by the boundary on the fluid. Written in a vector formula we have:

$$\vb{f}^B+\vb{f}^F=0\,\quad\tag{1}$$

where $\vb{f}^F$ is the stress excerted by the fluid and $\vb{f}^B$ the stress caused by the boundary. 

We know the magnitude and direction of $\vb{f}$, from Chapter 9, thus we can write

$$\vb{f}^F=\vb{n}^{outward}\cdot\tb{\tau}\quad .\tag{2}$$

The normal vector of the rigid boundary is in opposite direction to $\vb{n}^{outward}$, thus we can solve for the stress the boundary is inducing on the fluid particle:

$$\vb{f}^B=-\vb{n}^{outward}\cdot \tb{\tau}=\vb{n}^{inward}\cdot \tb{\tau}\tag{3}$$

This looks like an obvious relationship. It is, yet it allows to formulate boundary conditions for the fluid particles being in contact with rigid (non-moving) boundaries.

## Fluid Stresses Between two Fluids

Between two immiscible fluids such as a oil and water or water and gas boundary stresses have to be balanced. In contrast to a rigid boundary, fluid boundaries can deform. When a fluid boundary deforms from a plane to a curved shape more area is created. To create this additional energy a force needs to be applied. Thus a fluid-fluid interface does not simply balance stresses throught the surface normal as done in Eq. (1) but we need to account for additional forces along the interface. These forces are acting tangentially and we quantify the net tension with $\gamma\,\vb{t}$ where $\gamma$ is the interfacial tension coefficient and $\vb{t}$ the tangential vector.


The interfacial tension coefficient $\gamma$ is related to the "attractiviness" of the liquid. A molecule in a liquid "L" is surrounded by equal neighbours while a molecule at the interface between liquid and gas "G" has less partners. This preference of a molecule being surrounded by chemically similar molecules leads to a surface energy: one needs to do work on a liquid to form a new surface, or in other words, the liquid wants to form an object of minmum surface area.

The surface energy can ge approximated by the binding forces of molecules. Here we estimate that energy is of the order of $U \approx k_b T\approx 1.38\cdot 10^{-23}\,$m$^2\,$kg$\,$s$^{-2}\,$K$^{-1}\, 300\,$K and the diameter of the water molecule about 2.9$\,$A. Thus the energy per surface is of the order of $U/(2 \cdot 2.9\,$A$)^2\approx 0.024$J$\,$m$^{-2}$ or $0.024$N$\,$m$^{-1}$. The energy per area or force per length (same units!) are not correct for liquid with a strong coehesion; for water $\gamma\approx 0.072\,$N$/$m$^2$.

The force is acting in the direction to reduce the surface area under certain constrains.



## Derivation of the Stress Balance



Let's write this balance formally for a one-dimensional fluid-fluid interface presented in Fig. 2

<div class="Figure">
<img src="files/pics/interfacebalance.png" width=450px> 
Figure 3:
</div>

The stress caused by the boundary is the inward normal vector times the stress tensor of the fluid element, thus the force (per length $\Delta l$) for fluid 1 is $(n^{(1)}\cdot\tau^{(1)})\Delta l$ and similar or fluid 2. We need also to account for the forces in tangential direction. Assuming that the tangential vector is pointing to the left we have the force to the left $\gamma^B\vb{t}^B$ and to the right $-\gamma^A\vb{t}^A$, where the minus sign is caused by the force at $A$ being directed into the line element $\Delta l$. From geometry $\vb{n}^{(1)}=-\vb{n}^{(2)}$ and we can write a force balance

$$(\vb{n}^{(1)}\cdot \tb{\tau}^{(1)})\Delta l + (\vb{n}^{(2)}\cdot \tb{\tau}^{(2)})\Delta l - \gamma^A\vb{t}^A + \gamma^B\vb{t}^B = \vb{0}\tag{2}$$

This can be rewritten as

$$\vb{n}^{(1)}\cdot\left(\tb{\tau}^{(1)}-\tb{\tau}^{(2)}\right)=-
\frac{\gamma^B \vb{t}^B-\gamma^A \vb{t}^A}{\Delta l}$$

for which we can formulate the limit $\lim_{\Delta l \rightarrow 0}$ such that

$$\vb{n}^{(1)}\cdot\left(\tb{\tau}^{(1)}-\tb{\tau}^{(2)}\right)=-\fud{\left(\gamma \vb{t}\right)}{l}=-\gamma \fud{\vb{t}}{l}-\vb{t}\fud{\gamma}{l}\tag{3}$$

We obtain two interesting terms on the right hand side. The first is altered when the surface is bend while the second is important when the surface tension is varying along the surface. The latter term is called the Marangoni effect.

With no derivation of the expression (available in Pozrikidis or at [Wikipedia](https://en.wikipedia.org/wiki/Curvature) we can write

$$\fud{\vb{t}}{l}=-\kappa \,\vb{n^{(1)}}\tag{4}$$ 

where $\kappa$ is the curvature of the interface. Thus Eq. (3) becomes

$$\Delta \vb{f}=\vb{n}^{(1)}\cdot\left(\tb{\tau}^{(1)}-\tb{\tau}^{(2)}\right)=\gamma \kappa \vb{n}^{(1)} - \vb{t}\fud{\gamma}{l}\tag{5}$$


The curvature of a circle with radius $R$ is $\kappa=\frac{1}{R}$. Thus we can consider our interface to be made up of pieces of circles with varying curvature. A flat interface has a radius of curvature of $R=\infty$ thus a curvature of $\kappa=0$. This leads to no interfacial stress due to the bending of the interface. The physical meaning behind the bending is that a curved surface posseses a larger surface area and therefore is in an energetical unfavourable state. Surface tension leads to a minimizing of the surface area. 

How doe we calculate the curvature of an arbitrary surface? Therefore we would need to apply some differential geometry. This is not covered in the course but can be relatively easily derived (see Pozrikids and Kundu & Cohen). The curvature for a plane curve where $y=f(x)$ is 

$$\kappa=-\frac{f''}{\left(1+(f')^2\right)^{3/2}}=\frac{1}{f'}\left(\frac{1}{\sqrt{1+(f')^2}}\right)'\tag{6}\quad ,$$
where $f'=\dd f / \dd x$.

## Young's Relationship

The three phase contact line is the line where three phases (solid, liquid, and/or vapour/gas) come into contact. This is the place where a liquid wets a solid for example. The contact can be described macroscopically with a line where three interfacial forces interact, each of them characterized by a coefficient of surface tension. The result is a contact angle which is the angle between the solid and the liquid. Have a look at Fig. 4, on the top you see a photograph of a drop wetting a substrate. The contact angle (measured within the liquid) is approximately $15^\circ$.

On the bottom of Fig. 4 the definition of the 3 surface tension coefficients are given: $\gamma_{LG}$ is the surface energy between liquid and gas, $\gamma_{SG}$ between solid and gas, and $\gamma_{SL}$ between solid and liquid. As the droplet is not moving, the resulting forces must balance to 0, thus we can derive the simple relationship for the forces in the horizontal

$$
\gamma_{SG}-\gamma_{SL}=\gamma_{LG} \cos \theta_C\tag{7}\quad .
$$

Note that the forces in the vertical also balance and the upward force $\gamma_{LG}\sin \theta_C$ is balanced by the elasticity of the substrate.

If only two phases are present the surface tension coefficient is written without the subscript, i.e. $\gamma$.


## Capillary Length

By comparing the hydrostatic pressure with the stress at an interface between two fluids we can define a length at which both pressures are of equal magnitude. The pressure jump from surface tension is $p_S=\gamma/l$ and the hydrostatic pressuee is $p_H=\rho g l$. Equating both pressures and solving for $l$ we obtain:

$$l=\sqrt{\frac{\gamma}{\rho g}}\tag{8}$$

Let's insert typical values for water and water vapour and we obtain $l\approx 2.6\,$mm. That is the length of a typical miniscus, over which surface tension smoothenes the surface in Earth's gravitational potential.

## Balancing Gravity with Surface Tension

The shape of drops and thin films are governed by the interplay between gravity and surface tension. An equation named after <a href="https://en.wikipedia.org/wiki/Thomas_Young_(scientist)"> Thomas Young </a> and [Pierre-Simon Laplace](https://en.wikipedia.org/wiki/Pierre-Simon_Laplace) connects the surface tension with the hydrostatic pressure change.

The jump in stress between side (2) and side (1) is given by the left side of Eq. (5). For a static liquid the stress is only due to pressure, i.e. $\tb{\tau}=-p \, \tb{I}$ with $\tb{I}$ being the identity matrix. Assuming a constant surface tension coefficient on the along the line in Figure 5 we can write Eq. (5) as

$$p^{(2)}-p^{(1)}=\gamma \kappa\tag{9}\quad .
$$

With the coordinate system aligned with the y-axis in opposite direction to the gravity vector (pointing downwards) the hydrostatic pressure $p$ is

\begin{eqnarray*}
p_1&=&-\rho_1 g y_1 + c_1\\
p_2&=&-\rho_2 g y_2 + c_2 \quad .
\end{eqnarray*}

Inserting above expressions into Eq. (9) and identifying $y_I=y_1=y_2$ gives

$$-\rho_2 g y_I + c_2 +\rho_1 g y_I - c_1 = \gamma \kappa\tag{10}\quad .$$

With $\Delta \rho=\rho_2-\rho_1$ and $B=(c_2-c_1)/\gamma$ Eq. (10) can be rewritten to

$$\kappa=-\frac{\Delta \rho \, g}{\gamma} y_I +B$$

and with the definition of the capillary length $l$ we obtain Young-Laplace equation

$$\kappa=-\frac{y_I}{l^2}+B\tag{11}\quad .$$
