# Conservation of Mass

## Gaussian divergence theorem


Consider a vector field $\vb{h}(\vb{x})$

$$\vb{h}=\begin{pmatrix}h_1(x_1,x_2)\\h_2(x_1,x_2)\end{pmatrix}\tag{1}$$

and the component in normal direction is given by 

$$h_n=\vb{h}\cdot\vb{n}=h_1 n_1 + h_2 n_2\tag{2}$$

with the definition of the divergence

$$\nabla \cdot \vb{h}= \pad{h_1}{x_1} + \pad{h_2}{x_2}\tag{3}$$

then the Gaussian divergence theorem states that the integral over the closed loop aong contour $\cal C$ is 

$$\oint_{\cal C} \vb{h}\cdot\vb{n} \,\dd l=
\int \int_{\cal D} \nabla \cdot \vb{h} \, \dd x_1 \, \dd x_2$\tag{4}\quad .$

The Gaussian divergence theorem is very helpful in formulating conservation laws. This becomes immediately visible if we use the flow field $\vb{u}$ as the vector field $\vb{h}$.


Then 

$$
Q=\oint_{\cal C} \vb{u}\cdot\vb{n} \,\dd l =
\int \int_{\cal D} \nabla \cdot \vb{u} \,\dd x_1 \,\dd x_2\tag{5}$$

The divergence operator is the sum of the trace of the velocity gradient tensor. We have identified this with linear strain. When the sum of the linear strain is zero the fluid particle does not expand or compress its area/volume.

Thus for incompressible flow

$$\nabla \cdot \vb{u}=0\tag{6}\quad .$$
