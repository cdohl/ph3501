# Conservation of Mass

## Gaussian divergence theorem


Consider a vector field $\vb{h}(\vb{x})$

$$\vb{h}=\begin{pmatrix}h_1(x_1,x_2)\\h_2(x_1,x_2)\end{pmatrix}\tag{1}$$

and the component in normal direction is given by 

$$h_n=\vb{h}\cdot\vb{n}=h_1 n_1 + h_2 n_2\tag{2}$$

with the definition of the divergence

$$\nabla \cdot \vb{h}= \pad{h_1}{x_1} + \pad{h_2}{x_2}\tag{3}$$

then the Gaussian divergence theorem states that the integral over the closed loop aong contour $\cal C$ is 

$$\oint_{\cal C} \vb{h}\cdot\vb{n} \,\dd l=
\int \int_{\cal D} \nabla \cdot \vb{h} \, \dd x_1 \, \dd x_2\tag{4}\quad . $$

The Gauss divergence theorem is very helpful in formulating conservation laws. This becomes immediately visible if we use the flow field $\vb{u}$ as the vector field $\vb{h}$.


Then 

$$
Q=\oint_{\cal C} \vb{u}\cdot\vb{n} \,\dd l =
\int \int_{\cal D} \nabla \cdot \vb{u} \,\dd x_1 \,\dd x_2\tag{5}$$

The divergence operator is the sum of the trace of the velocity gradient tensor. We have identified this with linear strain. When the sum of the linear strain is zero the fluid particle does not expand or compress its area/volume.

Thus for incompressible flow

$$\nabla \cdot \vb{u}=0\quad \tag{6}$$

which will support in a brief moment.

## Conservation of Mass

For the conservation of mass we need to define the mass flux $q_m=\rho u_n$ and the integral over a line/area is the mass flow rate

$$Q_m=\int_A^B q_m \dd l = \int_A^B \rho \vb{u}\cdot\vb{n} dl=
\int_A^B \rho \left(u_1 \dd x_2 - u_2 \dd x_1\right)\quad .\tag{6}$$ 

To apply the divergence theorem we need to form a closed surface area as shown in Fig. 1.

\begin{eqnarray}
Q_m&=&\oint_A^B q_m \dd l =\oint_A^B \rho \left(u_1 \dd x_2 - u_2 \dd x_1\right)\tag{8a}\\
&=& \int\int_{\cal D} \nabla \cdot \left(\vb{\rho}\vb{n}\right) \dd x_1 \dd x_2\tag{8b}\\
&=& \int\int_{\cal D} \left( \vb{u}\cdot\nabla \rho + \rho\nabla \vb{u}\right)\dd x_1 \dd x_2\tag{8c}
\end{eqnarray}

In Eq. (8b) we have applied the Gaussian divergence theorem and in Eq. (8c) the chain rule.

If the fluid is compressible, than not necessarily the mass of the domain $\cal D$ remains constant. The change of mass in domain $\cal D$ is equal to the mass flux. The mass is calculated by the area/volume integral of the density. Let us write this idea as a formula:

$$\fud{}{t}\int \int \rho \dd x_1 \dd x_2=-Q_m=-\int \int_{\cal D} \nabla \cdot (\rho \vb{u}) \,\dd x_1 \dd x_2\tag{9}$$

The domain $\cal D$ is fixed in space, therefore the time derivative does not affect the boundaries and Equation (9) can be written as a single kernel:

$$\int \int \pad{\rho}{t} +\nabla \cdot (\rho \vb{u})\,\dd x_1 \dd x_2=0\tag{10}\quad .$$

Now an interesting argument is used. The domain $\cal D$ has not specific properties besides being enclosed by a loop $\cal C$. As the domain has not particular properties and can be applied to any part of the flow, the kernel itself must be zero. Thus we can demand that 

$$\pad{\rho}{t} +\nabla \cdot (\rho \vb{u})=0\tag{11a}$$

or

$$\pad{\rho}{t} +\nabla \rho \cdot \vb{u} + \rho \, \nabla \cdot \vb{u}=0\tag{11b}\quad .$$

With the [definition](03_Material Derivative.ipynb) of the *material derivative* 
we can rewrite Eq. (11a) as 

$$\fudm{\rho}{t}+\rho\,\nabla\cdot\vb{u}=0\tag{12}\quad .$$

Looking at Eq. (12) the definition of incompressible flow is straight forward. An incompressible fluid particle is a fluid particle who's denisty does not change, i.e. $\fudm{\rho}{t}=0$. If this is zero, $\nabla \cdot \vb{u}=0$ must hold. Which is equivallent that the sum of the linear rates of strain is 0.
