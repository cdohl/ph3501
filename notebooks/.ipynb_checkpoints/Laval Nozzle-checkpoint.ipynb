# Kompressible Strömung in Rohre mit veränderlichem Querschnitt

## Grundgleichungen

Kontinuitätsgleichung in *statinonären Strömungen* besagt dass der Massenstrom in durch die Fläche 1 gleich dem Massestrom durch die Fläche 2 ist. 


<img src="lpics/Rohr.png" width=350px>

Das lässt sich formulieren als

$$\dot{m_1}=\dot{m_2}\tag{1}$$

wobei der Massenstrom $\dot{m}$ gegeben ist als

$$\dot{m}=\rho u A\tag{2}\quad .$$

Damit ist der Massenstrom an jeder Position $x$ identisch und

$$\dot{m}= \mbox{const.}$$

Wir wollen nun die Abhängigkeit von der Strömungsgeschwindigkeit $u$ von dem Rohrqueschnittsverlauf $A(x)$ betrachten. Dazu bilden wir das totale Differential des Massenstroms

$$\d m=\rho u\, \d A + \rho A\, \d u + u A\, \d \rho=0\quad .$$

und durch Teilen von beiden Seiten mit $\rho\, u\, A\, \d x$ erhalten wir

$$\frac{1}{A}\fud{A}{x}+\frac{1}{u}\fud{u}{x}+\frac{1}{\rho}\fud{\rho}{x}=0\tag{3}\quad .$$

Der letze Term in Gleichung (3) kann durch die Definition der Schallgeschwindigkeit 

$$a^2=\left(\pad{p}{\rho}\right)_S$$

ausgedrückt werden:

$$\frac{1}{A}\fud{A}{x}+\frac{1}{u}\fud{u}{x}+\frac{1}{\rho a^2} \fud{p}{x}=0\tag{4}$$

Wir wollen eine allgemeine Beziehung fuer die Geschwindigkeitsänderung, $u$, als Funktion der Querschnittsänderung $A$ herleiten. Da stört der Druckgradient in Gl. (4). Wir wissen aber, dass für eine reibungsfreie zeitunabhängige Gaströmung die Euler Gleichung 

$$\fudm{u}{t}=u\fud{u}{x}=-\frac{1}{\rho}\fud{p}{x}\tag{5}$$

gilt. Der Druckgradient $\d p/\d x$ kann somit aus Gleichung (4) mit einem Geschwingkeitsgradienten ersetzt werden und wir erhalten

$$\frac{1}{A}\fud{A}{x}=\frac{1}{u}\fud{u}{x}\left[M^2-1\right]\tag{6} \quad .$$

Die Zahl $M$ ist die Machzahl

$$M=\frac{u}{a}\tag{7}\quad .$$

Wir sprechen von Unterschallströmungen wenn $M<1$ und von Überschallströmungen mit $M>1$.

Betrachten wir Gleichung (6) können wir vier Fälle bequem mit Hilfe der Vorzeichen  unterscheiden.

## Strömungsgrössen

Für eine stationäre reibungsfreie Strömung gilt die Bernoulli Gleichung entlang einer Stromlinie (homentropes Gas, d.h. kein Stoss).

$$ \frac{1}{2} u^2 + P = \mbox{const.}\tag{8}$$

wobei die Druckfunktion durch 

$$P=\int \frac{1}{\rho (p)} \d p$$

gegeben ist. Wir nehmen ein ideales kalorisches Gas (ideales adiabtisches Gas mit konstantem $c_p$ & $c_v$) an: 
 
$$p=C\rho^\gamma\tag{9}$$


Dann ergibt sich für $P$

$$P=C^{1/\gamma}\frac{\gamma}{\gamma-1}p^{\frac{\gamma-1}{\gamma}}\tag{10}$$

Wir können die Konstante $C^{1/\gamma}$ in Gleichung (10) entweder durch einen zweiten Druck $p_1$ und Dichte $\rho_1$ bestimmenm, oder den ursprünglichen Druck $p$ und Dichte $\rho$, d.h. $C^{1/\gamma}=p_1^{1/\gamma}/\rho_1=p^{1/\gamma}/\rho$. Dann ergeben sich die folgenden beiden Druckfunktionen:

\begin{eqnarray}
P&=&\frac{\gamma}{\gamma-1}\frac{p_1}{\rho_1}\left(\frac{p}{p_1}\right)^{\frac{\gamma-1}{\gamma}}\tag{11}\\
P&=&\frac{\gamma}{\gamma-1}\frac{p}{\rho}\tag{12}
\end{eqnarray}

Damit können wir die Bernoulligleichung (8) für kompressible Strömungen schreiben

$$\frac{u^2_1}{2}+\frac{\gamma}{\gamma-1}\frac{p_1}{\rho_1}=
\frac{u^2_2}{2}+\frac{\gamma}{\gamma-1}\frac{p_1}{\rho_1}\left(\frac{p_2}{p_1}\right)^{\frac{\gamma-1}{\gamma}}\tag{13}$$

die für kalorische ideale Gase gilt.

Setzen wir die Position 1 von Gl. (13) in einen grossen Kessel ist die Geschwindingkeit $u_1\approx 0$, und $p_1$ und $\rho_1$ sind Kesseldruck und Dichte. Wenn das Gas in ein Gebiet mit $p_1$ expandiert ergibt sich für die Geschwindigkeit

$$u_2=\sqrt{\frac{2\gamma}{\gamma-1}\frac{p_1}{\rho_1}\left(1-\left(\frac{p_2}{p_1}\right)^{\frac{\gamma-1}{\gamma}}\right)}\tag{14}.$$

Die maximale Geschwindingkeit die Luft mit Normaldruck und Dichte bei Expansion ins Vakuum erreichen kann ist $u_2 \approx 730\,$m/s (mit $p_1\approx10^5\,$Pa, $\gamma=1.4$, $\rho_1=1.3\,$kg$\,$m$^{-3}$, und $p_2$=0:

$$u_2=\sqrt{\frac{2\gamma}{\gamma-1}\frac{p_1}{\rho_1}}\tag{15}.$$

## Ideales isentropes Gas in der Lavaldüse

Wir wollen ein Beschreibung der Grössen $a$, $p$, $\rho$, und $T$ in der Lavaldüse erzielen. 

Aus der Definition der Schallgeschwindigkeit folgt aus der Isentropennbeziehug: 

$$a^2=\left( \pad{p}{\rho}\right)_S =\gamma\,\frac{p}{\rho}\tag{16}=\gamma {\cal R}_s T\tag{16}$$

die wir nun in die Bernoulli Gleichung (13) auf der linken Seite einsetzen und die rechte Seite mit der Ruhegrösse $a_t$ identifizieren.

$$\frac{u^2}{2}+\frac{1}{1-\gamma} a^2 = \frac{1}{1-\gamma} a_t^2\tag{17}$$

Wegen Gleichung (16) ist das Verhältnis der Schallgeschwindigkeit mit der Temperatur 

$$\frac{T_t}{T}=\left(\frac{a_t}{a}\right)^2=1+\frac{\gamma - 1}{2} M^2\tag{18}$$


```python
#standard boilerplate for Python
%matplotlib inline 
import numpy as np
import matplotlib.pylab as plt
import math

#This is our velocity function
def velocity(x1, x2, t):
    omega = 1
    g = 0.5
    u1 = g*x1-omega*x2
    u2 = -g*x2+omega*x1

    return u1, u2
```

Now we put the Euler Method to work:

```python
nt = 10000 #number of steps
deltat = 0.002 #step size

x10 = 0.1 #initial position 
x20 = 0.1
x1 = np.zeros(nt) #thats where we store the result
x2 = np.zeros(nt)
x1[0] = x10  #insert the initial conditions
x2[0] = x20

for n in range(nt-1):
    t = deltat*n
    u1, u2 = velocity(x1[n],x2[n],t) #get velocity at that position 

    #This is the Euler step
    x1[n+1] = x1[n]+deltat*u1 #update the position
    x2[n+1] = x2[n]+deltat*u2

plt.plot(x1,x2);
plt.xlabel('$x_1$'),plt.ylabel('$x_2$');
```
