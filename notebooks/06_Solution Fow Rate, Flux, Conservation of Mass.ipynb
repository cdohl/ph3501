#Flow rate & Flux

Here we will find a general concept to quantify the amount of flow through a certain region. The volumetric flow rate is the amount of volume flowing per time through an area. This area could be the flow leaving a pump. The volumetric flow rate has as unit $L^3 T^{-1}$, i.e. volume per time. Below we will use 2d flow, therefore we are interested of the flow through a line, the flow rate is therefore in units $L^2 T^{-1}$ and named the areal flow rate. 
To get a clear definition of the flow rate we make use of below figure.
<img src="files/pics/flowrate_fig1.png" width=300> <p> 
The flow rate accounts only the flow through the line, not along the line. The amount of flow over a line element is therefore $u_n \, dl$ where $dl$ is a length of the line segment. $u_n$ is a short form for writing the velocity component normal to the line element $dl$. We can easily calculate this component by taking the scalar product of the the velocity field with the normal vector, $u_n=\vb{u} \cdot \vb{n}$. Only the normal component contributes to the flow through the line $dl$. We now need to calculate the normal vector to a parameterized curve, say $\vb{x}(l)$. The tangential vector $\vb{t}=t_1 \vb{e}_1 +t_2 \vb{e}_2$ clearly is 

$$\begin{eqnarray}
t_1&=&\fud{x_1}{l}\\
t_2&=&\fud{x_2}{l} 
\end{eqnarray}
$$

The normal component can be calculated from 

$$\begin{eqnarray}
n_x&=&\fud{x_2}{l}\\
n_y&=&-\fud{x_1}{l} 
\end{eqnarray}
$$

You can easily show that $\vb{t}\cdot\vb{n}=0$. The velocity field along the line can be written as $\vb{u}=u_n\vb{n}+u_t\vb{t}$. 

The local flow rate is the product of the normal velocity $u_n$ times the line element $dl$:

$$
\fud{a}{t}\equiv u_n dl\quad,
$$

while the local areal flux (flow rate per length)

$$
q\equiv\frac{\fud{a}{t}}{dl}=u_n \quad.
$$

We can now express the local areal flow rate in cartesian coordinates:

\begin{eqnarray*}
\fud{a}{t}=u_n d_l=\vc{u} \cdot \vc{n} dl&=&\left(u_1 n_1 + u_2 n_2\right) dl\\
&=&
u_1 dx_2 - u_2 dx_1
\end{eqnarray*}

The total areal flux through the line is then the integral of the local areal flux:

$$Q=\int_A^B q\,dl=\int_A^B \fud{a}{t}=\int_A^B u_1 dx_2 - \int_A^B u_2 dx_1$$

<img src="files/pics/closed_loop_fig2.png" width=300> <p> 

If the line is closed than above integral is that over a closed loop $C$ of domain $D$:

$$Q=\oint_C u_1 dx_2 - \oint_C u_2 dx_1$$

##Example

In this example we calculate the flow rate and flux generated by a point source. A point source is a infintely small region in a fluid domain which injects fluid homogeneously (radially). 

<img src="files/pics/example_pointsource_fig3.png" width=250> <p> 

If the flow is

\begin{eqnarray}
u_r&=& \frac{\alpha}{2\pi}\frac{1}{r}\\
u_\theta&=& 0 
\end{eqnarray}

we only deal with a radial component which decays with $1/r$. The constant factor will become clear at the end of this example. Let us now calculate the flux through a circular line segment with starting angle $\theta_A$ and $\theta_B$. We write above integrals written in cartesian coordinates into a planar cylindrical coordinate system. First we convert $u_1$ and $u_2$

\begin{eqnarray}
u_1&=&\frac{\alpha}{2 \pi}\cos \theta\\
u_2&=&\frac{\alpha}{2 \pi}\sin \theta
\end{eqnarray}

With $x=R\cos\theta$ and $y=R\sin\theta$ the variable of integrations can be changed to $d\theta$ and $R=$const., i.e. $dx=-R\sin\theta \, d\theta$ and $dy=R\cos\theta \, d\theta$. Inserting these expressions into the equation for the areal flux $Q$ we obtain

$$Q=R\int_A^B \left(\frac{\alpha}{2\pi R}\cos^2 \theta+\frac{\alpha}{2\pi R}\cos^2 \theta\right) d\theta$$
which is
$$Q=\frac{\alpha}{2 \pi}\left(\theta_A-\theta_B\right)\quad .$$

This results shows that the total areal flux is indendent on the distance from the source $R$. By increasing the distance $R$ the line segment becomes longer balancing the decrease in the normal fluid velocity. The result for a close curve is

$$Q=R\oint .... d\theta=R\int_0^{2\pi}... d\theta=\alpha$$
where the 3 points stand for above expression. Now the definitio of the constant $\alpha$ is becoming clear, it is the source strength of the radial flow.
