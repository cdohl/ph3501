```python
#standard boilerplate for Python
%matplotlib inline
from matplotlib.pyplot import cm
import matplotlib.pylab as plt
from IPython import display
import time

nn=0


def plot(u):
    global nn
    plt.figure(1, figsize=(10, 6), dpi=200)

    display.clear_output(wait=True)

    plt.clf()
    nn +=1
    plt.plot(u,'o',markersize=15, color='#808080')
    plt.box
    plt.axis([0,len(u)-1,0,6])
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.text(1,5,'Iteration {0}'.format(nn-1),size=15)
    plt.show()
    if nn==1:
        time.sleep(20)
    else: 
        time.sleep(.5)


u=[1,0,0,0,0,0,0,0,5]

plot(u)

for n in range(20):

    for i in range(1,len(u)-1):

        u[i]=1./2.*(u[i-1]+u[i+1])
    
    plot(u)
```
